## What are the top 30 Disny Movies with Highest Rotten Tomatoes Score ?

```python
rotten = df.dropna(axis=0, subset=['rotten_tomatoes','Title'])
rotten1=rotten[['Title','rotten_tomatoes']]
rotten1['rotten_tomatoes']=rotten1['rotten_tomatoes'].str.rstrip("%").astype(int)
rotten2=rotten1.sort_values(by=['rotten_tomatoes'],ascending=False).head(30)
rotten2
```

```python
plt.figure(figsize=(18,9))
plt.title('Top 30 Disny Movies with Highest Rotten Tomatoes Score ', size = 20)
sns.barplot(data=rotten2, y='Title', x ='rotten_tomatoes',linewidth = 2,palette='viridis')
plt.ylabel("Movie's Title", size=10)
plt.xlabel('Rating',size=10)
plt.show()
```

## What are the top 30 Disny Movies with Lowest IMDB Score ?

```python
imdb = df.dropna(axis=0, subset=['imdb','Title'])
imdb=imdb[['Title','imdb']]
imdb['imdb']=imdb['imdb'].astype(int)
imdb2=imdb.sort_values(by=['imdb'],ascending=False).tail(30)
imdb2
```

```python
plt.figure(figsize=(10,5))
plt.title('Top 30 Disny Movies with Lowest IMDB Score ', size = 20)
sns.barplot(data=imdb2, y='Title', x ='imdb',linewidth = 2)
plt.ylabel("Movie's Title", size=10)
plt.xlabel('Rating',size=10)
plt.show()
```

## Top 5 Disney Movies with the Highest Budget

```python
Budget_Fixed = df.dropna(axis=0, subset=['Budget_Fixed','Title'])
Budget_Fixed =Budget_Fixed [['Title','Budget_Fixed']]
Budget_Fixed['Budget_Fixed']=Budget_Fixed['Budget_Fixed'].astype(int)
Budget_Fixed2=Budget_Fixed.sort_values(by=['Budget_Fixed'],ascending=False).head(5)
Budget_Fixed2
```

```python
plt.figure(figsize=(20,9))
plt.title('Top 5 Budgets for Disny Movies', size = 20)
sns.barplot(data=Budget_Fixed2, y='Budget_Fixed', x ='Title',linewidth = 2)
plt.xlabel("Movie's Title", size=10)
plt.ylabel('Budget',size=10)
plt.show()

print(' The highest budget would be 410.6 million for Pirates of the Caribbean: On Stranger Tides !')
```

## Correlation between Budget and Rotten Tomatoes score

```python
corr = df.dropna(axis=0, subset=['Budget_Fixed','Title','rotten_tomatoes'])
corr =corr [['Title','Budget_Fixed','rotten_tomatoes']]
corr['Budget_Fixed']=corr['Budget_Fixed'].astype(int)
corr['rotten_tomatoes']=corr['rotten_tomatoes'].str.rstrip("%").astype(int)
corr
```

```python
plt.figure(figsize=(15,6))
sns.regplot(data=corr, x='Budget_Fixed',y='rotten_tomatoes',scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.title('Budget vs Ratings', size=17)
plt.show()

print("Budget & Rotten Tomatoes's Ratings has no correlation of each other since the correlation is 0.08 which is nearly 0")
```
